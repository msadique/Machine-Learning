```python
%matplotlib inline
import numpy as np
str = "PNPPPNPNNPNN"
P=6
N=6
p=0
n=0
tpr =[]
fpr =[]
for s in str:
    if(s=='P'):
#       print("P")
        p=p+1
    else:
#        print("N")
        n=n+1
    tpr.append(p/P)
    fpr.append(n/N)
print(p,n)
TPR = np.array(tpr)
FPR = np.array(fpr)

print(TPR,"sadique",FPR)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pylab
pylab.rcParams['figure.figsize']= (12.0,12.0)
```

```python
t1 = []
f1 = []
for i in range (10):
    t1.append(i/6)
    f1.append(i/6)
T1 = np.array(t1)
F1 = np.array(f1)
fig = plt.figure()
ax = fig.gca()
ax.set_xticks(F1)
ax.set_yticks(T1)
plt.scatter(FPR,TPR)
plt.grid()
ax.set_title('Receiver Operating Characteristic (ROC) Curve ')

ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')
ax.plot(FPR, TPR, label='AUC')
ax.axhline(y=0, color='Red',label='X-axis')
ax.axvline(x=0, color='Green',label='Y-axis')
#ax.legend()

plt.show()
```
